```python
import pandas as pd
import numpy as np
import os, json
```

```python
sh = pd.read_csv('Data/superhero_info - superhero_info.csv')
sh.head()
```

```python
sp = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
sp.head()
```

```python
#Split Hero/Publisher into two columns
sh['Hero|Publisher'].head()
```

```python
sh['Hero|Publisher'].str.split('|', expand=True)
```

```python
sh[['Hero', 'Publisher']] = sh['Hero|Publisher'].str.split('|', expand=True)
sh.head()
```

```python
sh = sh.drop(columns = 'Hero|Publisher')
sh.head()
```

```python
col_order = ['Hero', 'Publisher', 'Gender', 'Race', 'Alignment', 'Hair color',
            'Eye color', 'Skin color', 'Measurements']

sh = sh[col_order].copy()
sh.head()
```

```python
#Split Measurements into height and weight
sh['Measurements'].apply(pd.Series)
```

```python
test_meas = sh.loc[1, 'Measurements']
test_meas
```

```python
type(test_meas)
```

```python
test_meas = test_meas.replace("'", '"')
fixed_meas = json.loads(test_meas)
print(type(fixed_meas))
fixed_meas
```

```python
sh['Measurements'] = sh['Measurements'].str.replace("'", '"')
sh['Measurements'] = sh['Measurements'].apply(json.loads)
sh['Measurements'].head()
```

```python
height_weight = sh['Measurements'].apply(pd.Series)
height_weight
```

```python
sh = pd.concat((sh, height_weight), axis=1)
sh.head()
```

```python
sh.drop(columns='Measurements', inplace=True)
sh.head()
```

```python
sh['Height'] = sh['Height'].str.replace(' cm', '')
sh['Height'] = sh['Height'].astype(float)
sh['Height']
```

```python
sh['Weight'] = sh['Weight'].str.replace(' kg', '')
sh['Weight'] = sh['Weight'].astype(float)
sh['Weight']
```

```python
#One hot encode columns for everey power that appears in the data set
sp.head()
```

```python
sp['Powers'].head()
```

```python
sp['Powers_Split'] = sp['Powers'].apply(lambda x: x.split(','))
sp['Powers_Split']
```

```python
exploded = sp.explode('Powers_Split')
exploded[['hero_names', 'Powers', 'Powers_Split']].head()
```

```python
cols_to_make = exploded['Powers_Split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    sp[col] = sp['Powers'].str.contains(col)

sp.head()
```

```python
cols_to_keep = ['hero_names', 'Powers']
sp_new = sp[cols_to_keep]
sp_new.head()
```

```python
#Join into one final pandas frame
merged = pd.merge(sh, sp_new, left_on = 'Hero', right_on = 'hero_names', how='inner')
merged.head()
```

```python
df = merged.drop(columns='hero_names')
df.head()
```

```python
speed = 'Speed'

with_super_speed = df[df['Powers'].str.contains(speed)]
with_super_speed
```

```python
no_super_speed = df[~df['Powers'].str.contains(speed)]
no_super_speed
```

```python
print(with_super_speed['Weight'].mean())
print(no_super_speed['Weight'].mean())
```

```python
#Heros with super speed, on average, weigh 129 kg
#Heros without super speed, on average, weigh 101 kg
```
